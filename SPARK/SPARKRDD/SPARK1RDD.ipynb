```python
import pyspark
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName('spark').getOrCreate()
```

```python
from google.colab import drive
drive.mount('/content/drive/')
#Read a text file
data=spark.sparkContext.textFile('sampleTemp.txt')
```

```python
data.collect()
```

```python
#Count the number of words in the file using RDD operations.
dataFlatMap=data.flatMap(lambda d:d.split(' '))
dataFlatMap.count()
```

```python
#Write a program to find the word frequency in a given file
dataFlatMap.map(lambda d:(d,1)).reduceByKey(lambda a,b:a+b).collect()
```

```python
#Write a program to convert all words in a file to uppercase
def isAlphabet(d):
  if d.isalpha():
    return True
dataFlatMap.filter(lambda d:isAlphabet(d)).map(lambda d:d.upper()).collect()
```

```python
#Write a program to convert all words in a file to lowercase.
def isAlphabet(d):
  return d.isalpha()
dataFlatMap.filter(lambda d:isAlphabet(d)).map(lambda d:d.lower()).collect()
```

```python
#Write a program to capitalize first letter of each words in file
def isAlphabet(d):
  return d.isalpha()
dataFlatMap.filter(lambda d:isAlphabet(d)).map(lambda d:d.capitalize()).collect()
```

```python
#Find the number of occurrence of a word in a given file
dataFlatMap.map(lambda d:(d,1)).reduceByKey(lambda a,b:a+b).collect()
```

```python
#Select only the sentences containing given word from a text file
def patternCheck(d):
  return 'Abohar' in d
data.filter(lambda d:patternCheck(d)).collect()
```

```python
# Find the longest length of word from given set of words
dataFlatMap.map(lambda d:(len(d),d)).max()
```
